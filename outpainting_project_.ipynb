```python
from PIL import Image

# Load the original image
original_image = Image.open('/content/assignment.jpg')

# Create a new image with extended size
new_size = (1280, 1280)
extended_image = Image.new('RGB', new_size, (255, 255, 255)) # white background
extended_image.paste(original_image, (128, 128))

# Save the extended image to check
extended_image.save('/content/extended_image.jpg')
extended_image.show()
```

```python
import torch
from diffusers import StableDiffusionInpaintPipeline

# Load the model
model_id = "runwayml/stable-diffusion-inpainting"
pipe = StableDiffusionInpaintPipeline.from_pretrained(model_id)

# Check if CUDA is available and move the model accordingly
if torch.cuda.is_available():
    pipe = pipe.to("cuda")
else:
    print("CUDA is not available. Running on CPU.")


# Prepare the prompt and mask
prompt = "A natural extension of the image with a fantasy theme"
mask_image = Image.new('L', new_size, 0)
mask_image.paste(255, (0, 0, 1280, 128))  # Top
mask_image.paste(255, (0, 1152, 1280, 1280))  # Bottom
mask_image.paste(255, (0, 0, 128, 1280))  # Left
mask_image.paste(255, (1152, 0, 1280, 1280))  # Right

# Run the outpainting
result = pipe(prompt=prompt, image=extended_image, mask_image=mask_image).images[0]

# Save and display the result
result.save('/content/outpainted_image.jpg')
result.show()
```
